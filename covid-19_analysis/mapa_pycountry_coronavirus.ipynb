```
import pandas as pd
df = pd.read_csv('totalescv.csv')
df.head(5)
```

```
df = df.rename({'index':'paises',
                53:'total'},axis=1)
df
```

```
import pycountry
def get_alpha_3(location):
  try:
    return pycountry.countries.get(name=location).alpha_3
  except:
    return None
df['Code'] = df['paises'].apply(lambda x: get_alpha_3(x))
df
```
